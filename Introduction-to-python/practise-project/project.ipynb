```python
year_release = item['releaseDATE'].str.split('-', expand=True)[2]
item['release year'] = year_release
item.dropna(axis=0, subset=['release year'], inplace=True)
item
```

```python
genere_by_year = item.groupby('release year');
genere_by_year = genere_by_year.sum()
genere_by_year = genere_by_year.drop(columns = ['movieID']).T
```

```python
plt.figure(figsize=(20,7))
sns.heatmap(genere_by_year, cmap='YlGnBu');
```

```python
top_movies = data.groupby('itemID').count()
top_movies = top_movies[top_movies.userID >100]
indexes = top_movies.index
top_movies = data[data.itemID.isin(indexes)]
top_movies = top_movies.groupby("itemID").mean()
top_movies = top_movies.sort_values('rating', ascending=False)
order = top_movies.index
rating = top_movies.rating[0:25]
```

```python
top_movies = item.loc[item['movieID'].isin(order)]
top_movies = top_movies.set_index('movieID')
top_25_movies = top_movies.iloc[0:25, 0]
top_25_movies = top_25_movies.reset_index()
top_25_movies['avg_rating'] = rating.values
```
